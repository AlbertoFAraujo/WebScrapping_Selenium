```python
# Atualização do WebDrive for Chrome de forma automatica
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

# Componentes webdriver para selecionar os elementos em (find_elements(By.))
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# Caso necessário utilizar o time.sleep
import time

# Tratamento de dados
import pandas as pd
import numpy as np
```

```python
# Cria a instância do Chrome e exibe a tela
servico = Service(ChromeDriverManager().install())
nav = webdriver.Chrome(service=servico)
```

```python
# Acessar o site AASP:
site = 'https://www.aasp.org.br/suporte-profissional/indices-economicos/mensal/dolar/'
nav.get(site)
```

```python
# Acessar os dados de histórico do dólar:
hist_dolar = nav.find_element(By.XPATH, '/html/body/div[1]/div[3]/div/div/div/div/figure').text
print(hist_dolar)
```

### Data Wrangling

Será necessário realizar a limpeza dos dados, transformá-los em uma estrutura adequada para tratamento, limpeza e organização.

```python
# Realizando a divisão das linhas (Note que o caracter \n divide as linhas)
linhas = hist_dolar.split('\n')
dados_lista = [linha.split() for linha in linhas]
dados_lista
```

```python
df = pd.DataFrame(dados_lista)
display(df)
```

```python
# Selecionar a primeira linha do próprio df como sendo o cabeçalho
df.set_axis(df.iloc[0], axis=1, inplace=True)
df = df[1:]
df
```

```python
# Alterando os nomes dos cabeçalhos
df = df.rename(columns={'Ano/Mês':'Ano','Março':'Marco'})
df.head(5)
```

```python
# Alterar o formato dos valores (Vírgula > Ponto)
df = df.replace(',','.', regex=True)
display(df.head(10))
```

```python
# Verificando o tipo de dado das demais variáveis
df.dtypes
```

```python
# Verificando as últimas observações
df.tail(5)
```

```python
# Excluíndo a última linha
df = df.drop(55, axis=0)
```

```python
# Filtrando a base até o ano de > 1994
linha_ref = df[df['Ano'] == '1994'].index[0] - 1

df = df.iloc[:linha_ref,:]
```

```python
df.tail(5)
```

```python
# Transformando os dados em float
df = df.astype(np.float32, copy = False)
df.dtypes
```

```python
# Alterando o tipo de dado da variável 'Ano' para inteiro
df['Ano'] = df['Ano'].astype(np.int64, copy = False)
df.dtypes
```

```python
# Alterando a quantidade de casas decimais
df = round(df,ndigits=2)
df.head(5)
```
